```python
import unicodedata
import re
import itertools
# from collections import defaultdict # No longer needed within preprocess_token itself

# (Keep your symbol_mapping and transliterate_greek functions as they were)

def normalize_all_unicode(text):
    text = unicodedata.normalize("NFKD", text)
    return ''.join(c for c in text if not unicodedata.combining(c))

def preprocess_token(token):
    """
    Preprocesses a single token by applying a series of normalizations and
    generating multiple interpretations, especially for bracketed text.

    Returns a list of potential string forms for the token.
    """
    # Step 1: Unicode normalization (e.g. 𝐍 → N)
    token = normalize_all_unicode(token)

    # Step 2: Replace smart quotes, dashes, etc.
    token = token.replace("’", "'").replace("‘", "'").replace("“", '"').replace("”", '"')
    token = token.replace("–", "-").replace("—", "-")  # Normalize dashes

    # Step 3: Apply symbol mapping
    for symbol, replacement in symbol_mapping.items():
        token = token.replace(symbol, replacement)

    # Step 4: Transliterate Greek (if any)
    token = transliterate_greek(token)

    # Step 5: Lowercase *after* all replacements
    token_lower = token.lower()

    # Step 6: Handle bracket-inclusive interpretations
    # This part needs careful design for complex bracket nesting/multiple brackets.
    # We will build `potential_parts` as a list of lists, where each inner list
    # contains the possible interpretations for that segment of the token.

    potential_parts = []
    current_segment = [] # To accumulate characters outside of brackets
    idx = 0

    while idx < len(token_lower):
        if token_lower[idx] == '(':
            # If we have accumulated a segment before the bracket, add it
            if current_segment:
                potential_parts.append([''.join(current_segment)])
                current_segment = [] # Reset for next segment

            end_idx = token_lower.find(')', idx)
            if end_idx != -1:
                inside_content = token_lower[idx+1:end_idx]
                if re.fullmatch(r"[a-z]+", inside_content): # Ensure it's purely alpha
                    potential_parts.append(['', inside_content]) # Empty string for exclusion, content for inclusion
                else:
                    # If invalid content in brackets, treat it as just the surrounding characters,
                    # effectively just ignoring the bracketed content.
                    potential_parts.append(['']) # Allow an empty choice if content is invalid
                idx = end_idx + 1 # Move past the closing bracket
            else:
                # Unmatched opening bracket, treat it as a literal character
                current_segment.append(token_lower[idx])
                idx += 1
        else:
            # Accumulate non-bracketed characters
            current_segment.append(token_lower[idx])
            idx += 1

    # Add any remaining accumulated segment after the loop
    if current_segment:
        potential_parts.append([''.join(current_segment)])

    # Now, combine these parts using itertools.product
    # Example: [['he'], ['', 'a'], ['t']] -> ('he', '', 't'), ('he', 'a', 't')
    
    all_outputs = set()

    if not potential_parts: # Handle empty or no-content tokens
        if token_lower and re.fullmatch(r"[a-z]+", token_lower):
            all_outputs.add(token_lower)
        return list(all_outputs)

    # Generate all combinations by taking one element from each sublist in potential_parts
    for combo in itertools.product(*potential_parts):
        full_word = "".join(combo)
        if full_word: # Ensure not empty string
            all_outputs.add(full_word)

    # Final filtering for valid tokens (lowercase alphabetic, non-empty)
    final_outputs = [t for t in all_outputs if t and re.fullmatch(r"[a-z]+", t)]

    # If no valid forms produced, and the original lowercased token is valid, use it as fallback.
    # This prevents returning an empty list for simple, valid tokens without brackets.
    if not final_outputs and token_lower and re.fullmatch(r"[a-z]+", token_lower):
        final_outputs.append(token_lower)
    elif not final_outputs:
        return [] # Return empty list if no valid forms can be derived at all

    return list(set(final_outputs)) # Return as a list of unique items
```

```python

import networkx as nx
import unicodedata
import re
import itertools
from collections import defaultdict

def preprocess_token(token):
    processed = normalize_all_unicode(token)
    processed = processed.replace("’", "'").replace("‘", "'").replace("“", '"').replace("”", '"')
    processed = processed.replace("–", "-").replace("—", "-")
    for symbol, replacement in symbol_mapping.items():
        processed = processed.replace(symbol, replacement)
    processed = transliterate_greek(processed)
    processed = processed.lower()
    alphabetic_chars = "".join([c for c in processed if c.isalpha()])
    if alphabetic_chars:
        return {alphabetic_chars}
    return set()


# --- Main graph building function (updated) ---
def build_word_level_graph(poem_text, word_list):
    G = nx.MultiDiGraph() 
    
    processed_word_list_set = set()
    for word in word_list:
        forms = preprocess_token(word)
        processed_word_list_set.update(forms)

    poem_word_sequence = []
    word_frequencies = defaultdict(int)

    lines = poem_text.splitlines()
    for line_idx, line in enumerate(lines):
        raw_words_in_line = re.findall(r"\S+", line) 

        for raw_poem_word in raw_words_in_line:
            forms_of_poem_word = preprocess_token(raw_poem_word)
            
            if not forms_of_poem_word:
                continue
            
            normalized_poem_word = next(iter(forms_of_poem_word)) 

            if normalized_poem_word not in G:
                is_from_word_list = normalized_poem_word in processed_word_list_set
                G.add_node(normalized_poem_word, 
                           text=normalized_poem_word, 
                           occurrence_count=0, 
                           frequency=0, 
                           label=normalized_poem_word, 
                           is_word_from_list=is_from_word_list) 
                
            G.nodes[normalized_poem_word]['occurrence_count'] += 1
            word_frequencies[normalized_poem_word] += 1 

            poem_word_sequence.append((normalized_poem_word, line_idx))

    # 3. Add Proximity Edges (Word-to-Word) - MODIFIED HERE
    for i in range(len(poem_word_sequence) - 1):
        word1, line1 = poem_word_sequence[i]
        word2, line2 = poem_word_sequence[i+1]

        if line1 == line2: # Within the same line
            G.add_edge(word1, word2, type="proximity", weight=1)
        elif line2 == line1 + 1: # Across an adjacent line break
            # This covers the end-of-line to beginning-of-next-line adjacency
            G.add_edge(word1, word2, type="proximity_line_break", weight=1)

    # 4. Add Reconstruct Edges (Word-to-Word Concatenation, Conceptual)
    existing_reconstruct_edges = set() 

    graph_nodes_as_words = list(G.nodes()) 

    for word_A in graph_nodes_as_words:
        for word_B in graph_nodes_as_words:
            if word_A == word_B: 
                continue
            
            combined_word = word_A + word_B
            
            if combined_word in processed_word_list_set:
                edge_signature = (word_A, word_B, "reconstruct")
                
                if edge_signature not in existing_reconstruct_edges:
                    # Defensive node creation (should already exist from Step 2)
                    if word_A not in G: 
                        G.add_node(word_A, text=word_A, occurrence_count=0, frequency=0, label=word_A, is_word_from_list=False)
                    if word_B not in G: 
                        G.add_node(word_B, text=word_B, occurrence_count=0, frequency=0, label=word_B, is_word_from_list=False)
                    
                    G.add_edge(word_A, word_B, type="reconstruct", weight=2)
                    existing_reconstruct_edges.add(edge_signature)

    # 5. Finalize Frequencies and Node Labels
    for node, data in G.nodes(data=True):
        G.nodes[node]['frequency'] = word_frequencies[node]
        G.nodes[node]['label'] = f"{node} ({G.nodes[node]['frequency']})"
        if G.nodes[node]['is_word_from_list']:
             G.nodes[node]['label'] += " [WL]" 
        else:
             G.nodes[node]['label'] += " [P]" 

    return G
```

```python

def normalize_graph_node_indices(graph):
    mapping = {node: i for i, node in enumerate(graph.nodes())}
    is_multi = graph.is_multigraph()
    is_directed = graph.is_directed()

    new_graph = (
        nx.MultiDiGraph() if is_multi and is_directed else
        nx.MultiGraph() if is_multi else
        nx.DiGraph() if is_directed else
        nx.Graph()
    )

    # Copy nodes
    for old_node, new_node in mapping.items():
        new_graph.add_node(new_node, **graph.nodes[old_node])

    # Copy edges properly, avoiding self-loops
    if is_multi:
        for u, v, key, data in graph.edges(keys=True, data=True):
            if u != v:  # prevent self-loops
                new_graph.add_edge(mapping[u], mapping[v], key=key, **data)
    else:
        for u, v, data in graph.edges(data=True):
            if u != v:
                new_graph.add_edge(mapping[u], mapping[v], **data)

    return new_graph
```

```python

def smart_sample_subgraphs(graph, num_subgraphs=100, subgraph_size=200, min_community_size=50):
    """
    Sample subgraphs from different communities in the graph,
    each including a balance of high, mid, low degree nodes.

    Args:
        graph: networkx Graph or MultiDiGraph
        num_subgraphs: total subgraphs to sample
        subgraph_size: target nodes per subgraph
        min_community_size: skip very tiny communities
    Returns:
        list of networkx subgraphs
    """
    # Use community detection to get structural clusters
    # communities = list(nx.algorithms.community.greedy_modularity_communities(graph.to_undirected()))
    communities = list(nx.algorithms.community.greedy_modularity_communities(graph))
    communities = [c for c in communities if len(c) >= min_community_size]

    # Flatten nodes to degrees for sampling diversity
    node_degrees = dict(graph.degree())

    subgraphs = []

    # Iterate over communities
    for community in communities:
        community_nodes = list(community)
        degrees = [node_degrees[n] for n in community_nodes]
        degree_bins = defaultdict(list)

        # Bin nodes by degree (high, mid, low)
        sorted_nodes = sorted(community_nodes, key=lambda n: node_degrees[n])
        n = len(sorted_nodes)
        for i, node in enumerate(sorted_nodes):
            if i < n * 0.33:
                degree_bins['low'].append(node)
            elif i < n * 0.66:
                degree_bins['mid'].append(node)
            else:
                degree_bins['high'].append(node)

        # Try multiple subgraphs from this community
        for _ in range(max(1, num_subgraphs // len(communities))):
            seeds = []
            if degree_bins['high']:
                seeds.append(random.choice(degree_bins['high']))
            if degree_bins['mid']:
                seeds.append(random.choice(degree_bins['mid']))
            if degree_bins['low']:
                seeds.append(random.choice(degree_bins['low']))

            # BFS to expand
            sub_nodes = set()
            for seed in seeds:
                neighbors = nx.single_source_shortest_path_length(graph, seed, cutoff=5).keys()
                sub_nodes.update(neighbors)
            sub_nodes = list(sub_nodes)[:subgraph_size]

            subgraph = graph.subgraph(sub_nodes).copy()
            subgraphs.append(subgraph)

            if len(subgraphs) >= num_subgraphs:
                break

        if len(subgraphs) >= num_subgraphs:
            break

    return subgraphs
```

```python

def graph_poem_to_text(graph):
    text_parts = []

    # --- Describe Nodes ---
    for n, data in graph.nodes(data=True):
        text = data.get("text", "")
        text_clean = text.replace("\n", " ").strip()
        # text_preview = text_clean[:120] + ("..." if len(text_clean) > 120 else "")  
        # desc = f"Node {n}: {text_preview}" if text else f"Node {n}: [No text]"
        desc = f"Node {n}: {text_clean}" if text else f"Node {n}: [No text]"
        text_parts.append(f"[Node {n}] {desc}")

    # --- Describe Edges ---
    for u, v, data in graph.edges(data=True):
        label = data.get("label", None)
        weight = data.get("weight", "")
        wildcard_delta = data.get("wildcard_delta", None)
        type_ = data.get("type", None)
        similarity = data.get("similarity", None)
        relation = data.get("relation", None)

        if label:
            label_preview = label.replace("\n", " ").strip()
            # label_preview = label_preview[:120] + ("..." if len(label_preview) > 120 else "")
            desc = f"Edge {u} → {v}: '{label_preview}'"
        elif data.get("type") == "semantic":
            sim = round(data.get("similarity", 0), 3)
            desc = f"Semantic Edge {u} → {v}, similarity={sim}"
        else:
            desc = f"Edge {u} → {v} (unlabeled)"

        # if wildcard_delta is not None:
        if wildcard_delta:
            delta_str = f", wildcard_delta={round(wildcard_delta, 4)}"
        else:
            delta_str = ""

        # if wildcard_cluster is not None:
        if type_ is not None and type_ == "wildcard_cluster":
            desc = f"{wildcard}"
            
        if similarity is not None:
            desc = f"{wildcard}"

        if relation is not None and relation == "context_similar":
            desc = f"{context}"
        
        weight_str = f", weight={weight}" if weight != "" else ""
        full_desc = f"{desc}{weight_str}{delta_str}{similarity}"
        text_parts.append(full_desc)

    return "\n".join(text_parts)
```

```python

from collections import defaultdict

def sort_node_pairs_by_edge_count(graph, top_n=5):
    # Count edges between node pairs
    edge_counts = defaultdict(list)

    for u, v, key, data in graph.edges(keys=True, data=True):
        edge_counts[(u, v)].append((key, data))

    # Sort pairs by number of connecting edges (descending)
    sorted_pairs = sorted(edge_counts.items(), key=lambda x: len(x[1]), reverse=True)

    for i, ((u, v), edges) in enumerate(sorted_pairs[:top_n]):
        print("=" * 60)
        print(f"Pair {i+1}: Node {u} ↔ Node {v} — {len(edges)} edge(s)")
        
        # Print Node u
        u_data = {k: v for k, v in graph.nodes[u].items() if k != "embedding"}
        print("\n- Node", u, "data:")
        for k, val in u_data.items():
            print(f"    {k}: {val}")

        # Print Node v
        v_data = {k: v for k, v in graph.nodes[v].items() if k != "embedding"}
        print("\n- Node", v, "data:")
        for k, val in v_data.items():
            print(f"    {k}: {val}")

        # Print all edge data between u and v
        print("\n- Edges between these nodes:")
        for key, edge_data in edges:
            edge_data_clean = {k: v for k, v in edge_data.items() if k != "embedding"}
            print(f"    Edge key {key}:")
            for k, val in edge_data_clean.items():
                print(f"        {k}: {val}")
```
